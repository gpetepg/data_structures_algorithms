<b>Terminology:</b>

<b>Root node:</b> At the top of every tree we have a root node. Ancenstor/parent to all the nodes in the tree.

<b>Node:</b> Represented here by the letters, a node is a structure that holds a value

<b>Sub-tree:</b> A section of the tree.

<b>Parent node:</b> A node that has children, e.g. D is the parent of H and I.

<b>Child node:</b> A node that has a parent, e.g. H and I are the children of D

<b>Leaf node:</b> A node with no children.

<b>Sibling:</b> Child nodes with the same parent.

<b>Height:</b> Total levels of the three, this tree has a height of 3.

<b>Level:</b> Levels to the tree, root node starts at 0, every time new children are introduced we add a level.

<b>Edge:</b> The connector between nodes.

<b>Depth:</b> The depth of a node is the number of edges from the root to the node.

<b>Categories of trees:</b>

<b>Full:</b> A full binary tree is a binary tree in which each node has exactly zero or two children.
    
<b>Complete:</b> A complete binary tree is a binary tree which is completely filled, with the possible exception of the bottom level, which is filled from left to right.

<b>Balanced:</b> When the differnece of height of subtrees are <= 1

<b>Fully balanced:</b> Full, complete, all nodes have two children and are the same height.

# Tree Traversal:

A systematic way to visit nodes in a tree, there are two types which have categories.

- <b>Depth-first (Stack)</b>
 - Pre-order
    1. Visit node
    2. Traverse left
    3. Traverse right
 - Post-order
    1. Traverse left
    2. Traverse right
    3. Visit node
 - In-order
    1. Traverse left
    2. Visit node
    3. Traverse right
- <b>Breadth-first (Queue)</b>
 - Level-order (Uses a queue)
    1. Enqueue root
    2. While the queue is not empty, peak then dequeue.
    3. If left or right child exist, enqueue left then right.
